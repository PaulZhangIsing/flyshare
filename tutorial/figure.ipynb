```python
import flyshare as fs  
import matplotlib.pyplot as plt

df = fs.get_hist_data('600519', start='2017-01-01', end='2017-01-24')

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
ax.set_xticks(range(0,len(df['date'])))
ax.set_xticklabels(df['date'])
ax.plot(df['open'])
ax.plot(df['close'])
ax.legend(loc='best')
fig.autofmt_xdate()
plt.grid()
plt.show()
```

```python
import talib
import numpy as np
import flyshare as fs
import matplotlib.pyplot as plt

data = fs.get_hist_data('399300', start='2017-01-01', end='2017-06-31')
sma_5 = talib.SMA(np.array(data['close']), 5)
sma_10 = talib.SMA(np.array(data['close']), 10)
sma_30 = talib.SMA(np.array(data['close']), 30)
cls = np.array(data['close'])

fig = plt.figure(figsize=(24, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks(range(0, len(data['date']), 10))
ax.set_xticklabels(data['date'][::10])
ax.plot(sma_5, label='5 day mean')
ax.plot(sma_10, label='10 day mean')
ax.plot(sma_30, label='30 day mean')
ax.plot(cls, label='close')
ax.legend(loc='upper left')
fig.autofmt_xdate()
plt.grid()
plt.show()
```
